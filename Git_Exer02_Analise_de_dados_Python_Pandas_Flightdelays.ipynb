```python
import pandas as pd
```

```python
# importando arquivos para dataframes 
fd1 = pd.read_csv('flightdelays/flight_delays1.csv')
# passando o parametro 'names' contendo a lista dos nomes das colunas do primeiro dataframe
fd2 = pd.read_csv('flightdelays/flight_delays2.csv', names=list(fd1.columns.values))
fd3 = pd.read_csv('flightdelays/flight_delays3.csv', names=list(fd1.columns.values))
```

```python
fd1.describe()
```

```python
fd2.describe()
```

```python
fd3.describe()
```

```python
# concatenando todos os datasets em um só
data_frame = pd.concat([fd1, fd2, fd3])
data_frame.describe()
```

```python
# removendo registros onde DepTime = 'NA'
# colunas: Year, Month, DayofMonth, DepTime, UniqueCarrier, FlightNum, ArrDelay, Origin and Dest
flightdelays = data_frame.loc[
    data_frame['DepTime'].isnull() != True, 
    ["Year", "Month", "DayofMonth", "DepTime", "UniqueCarrier", "FlightNum", "ArrDelay", "Origin", "Dest"]
]
display(flightdelays)
```

```python
# Parte 1
# Escreva um programa em Python que calcule o número de linhas do dataset flightdelays através de um dataframe no Pandas. 
# Armazene a saída do programa em um arquivo do tipo csv com o nome “clean_total”.
print("A quantidade de linhas do dataframe é igual a",flightdelays.shape[0])
flightdelays.to_csv('clean_total.csv')
```

```python
# Parte 2
# A coluna Dest é o código do aeroporto de destino onde o voo termina. 
# Escreva um programa em Python que calcula o número de linhas dos dados a partir do dataset flightdelays limpo 
# onde o campo Dest é igual ao código "DEN" do aeroporto de Denver. 
# Armazene a saída do seu programa em um arquivo csv chamado “denver_total”.
denver_total = flightdelays.loc[flightdelays['Dest'] == 'DEN']
print("A quantidade de linhas do dataframe é igual a",denver_total.shape[0])
display(denver_total)
denver_total.to_csv('denver_total.csv')
```

```python
# Parte 3
# A coluna ArrDelay é o número de minutos que um voo chegou atrasado. 
# Escreva um programa que conta o número de voos cujo destino é o aeroporto "DEN" 
# que chegaram com 60 ou mais minutos de atraso. 
# Armazene a saída do seu programa em um arquivo csv chamado “denver_late”
denver_late = flightdelays.loc[lambda df: (df['Dest'] == 'DEN') & (df['ArrDelay'] >= 60)]
print("A quantidade de linhas do dataframe é igual a",denver_late.shape[0])
display(denver_late)
denver_late.to_csv('denver_late.csv')
```
